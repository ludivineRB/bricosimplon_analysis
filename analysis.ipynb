```python
df_cat_centrale['categorie']=df_cat_centrale['categorie'].apply(unidecode.unidecode)
```

```python
df_cat_centrale
```

```python
#séparation en plusieurs df


# Catégories (niveau 1)
categories_df = df_cat_centrale[df_cat_centrale["type"] == "CAT"].rename(columns={
    "categorie": "Nom_Categorie",
    "id_categorie": "ID_Categorie"
})

# Sous-catégories (niveau 2)
subcategories_df = df_cat_centrale[df_cat_centrale["type"] == "SOUS_CAT"].rename(columns={
    "categorie": "Nom_Sous_Categorie",
    "id_categorie": "ID_Sous_Categorie",
    "id_cat_parent": "ID_Categorie_Parent",
    "lien_categorie": "Lien_Sous_Categorie"
})

# Sous-sous-catégories (niveau 3)
sous_sous_categories_df = df_cat_centrale[df_cat_centrale["type"] == "SOUS_CAT"].rename(columns={
    "categorie": "Nom_Sous_Sous_Categorie",
    "id_categorie": "ID_Sous_Sous_Categorie",
    "id_cat_parent": "ID_Sous_Categorie_Parent",
    "lien_categorie": "Lien_Sous_Sous_Categorie"
})
```

```python
# Relier les sous-catégories aux catégories parent
merged_subcategories = pd.merge(
    subcategories_df, 
    categories_df, 
    left_on="ID_Categorie_Parent", 
    right_on="ID_Categorie", 
    how="left"
).rename(columns={"Nom_Categorie": "Nom_Categorie_Parent"})
```

```python
# Relier les sous-sous-catégories aux sous-catégories parent
merged_sous_sous = pd.merge(
    sous_sous_categories_df, 
    subcategories_df.rename(columns={
        "Nom_Sous_Categorie": "Nom_Categorie_Sous_Categorie",
        "Lien_Sous_Categorie": "Lien_Categorie_Sous_Categorie",
        "ID_Sous_Categorie": "ID_Sous_Sous_Categorie_Parent"
    }),
    left_on="ID_Sous_Categorie_Parent", 
    right_on="ID_Sous_Sous_Categorie_Parent", 
    how="left"
)
```

```python
# Associer les catégories parent aux sous-sous-catégories via les sous-catégories
final_merged = pd.merge(
    merged_sous_sous, 
    merged_subcategories, 
    left_on="ID_Sous_Categorie_Parent", 
    right_on="ID_Sous_Categorie", 
    how="left"
)

# Filtrer les colonnes pertinentes
result = final_merged[[
    "Nom_Categorie_Parent", 
    "Nom_Sous_Categorie", 
    "Nom_Sous_Sous_Categorie", 
    "Lien_Sous_Sous_Categorie"
]]
```

```python
result.set_index(["Nom_Categorie_Parent", "Nom_Sous_Categorie", "Nom_Sous_Sous_Categorie"])
result
```
